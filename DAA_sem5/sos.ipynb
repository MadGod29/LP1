```python
def sum_of_subsets(arr, target_sum):
    n = len(arr)

    def subset_sum(curr_subset, curr_sum, index):
        if curr_sum == target_sum:
            print(f"Subset found: {curr_subset}")
            return
        
        if curr_sum > target_sum or index == n:
            return
        print(curr_subset, curr_sum, index)
        subset_sum(curr_subset + [arr[index]], curr_sum + arr[index], index + 1)
        print(curr_subset, curr_sum, index)
        subset_sum(curr_subset, curr_sum, index + 1)

    subset_sum([], 0, 0)

# Example usage
arr = [10, 7, 5, 18, 12, 20, 15]
target_sum = 35

sum_of_subsets(arr, target_sum)
```
